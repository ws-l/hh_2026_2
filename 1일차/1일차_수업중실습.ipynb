```python
import pandas as pd
from scipy import stats
df1 = pd.read_csv("experiment_01.csv")
df2 = pd.read_csv("experiment_02.csv")
x1 = df1.X_ActualVelocity
x2 = df2.X_ActualVelocity
stats.ttest_ind(x1, x2, equal_var=False)

before = [3,5,7,8,7,8,7,2,1] 
after = [3,5,6,10,9,10,7,4,1] 
stats.ttest_rel(before, after)
#유의수준 5%, 10%, 1%,
```

```python
import os
import pandas as pd
files = os.listdir("dataset\\dataset")
df = pd.DataFrame()
for i in files:
    tmp_df = pd.read_csv("dataset\\dataset\\"+i)
    tmp_df["source"] = i
    df = pd.concat([df, tmp_df])
df.describe()
df_num = df.drop(
    columns=["Machining_Process", "source"]
    )
df_num.corr()["X_ActualVelocity"].sort_values(ascending=False)
df[["X_CurrentFeedback","Y_ActualVelocity","Z_ActualVelocity"]].describe()


#......

from sklearn.preprocessing  import MinMaxScaler
from sklearn.cluster import KMeans
df_num_scaled = MinMaxScaler().fit_transform(df_num)
kc = KMeans(n_clusters=8).fit(df_num_scaled)
df["cl" ] = kc.labels_
kc.cluster_centers_
df.cl.value_counts()
```

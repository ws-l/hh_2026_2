#### 1. 카이제곱 적합도 검정

```python

from scipy.stats import chisquare

# 관측된 빈도
observed = [22,21,22,27,22,36]

# 기대되는 빈도 (모든 범주가 동일한 확률을 가질 경우)
expected = [25,25,25,25,25,25]

# 카이제곱 적합도 검정 수행
chi_square, p_value = chisquare(f_obs=observed, f_exp=expected)
print( chi_square, p_value)
```

#### 2. 카이제곱 독립성 검정

Type  | OperatorA |  OperatorB |  OperatorC

Gage1 | 37       |   41       |   44

Gage2 | 35       |   72       |   71

```python
from scipy.stats import chi2_contingency
import numpy as np

# 교차 테이블(Contingency table)
data = np.array([[37, 41, 44], [35, 72, 71]])

# 카이제곱 독립성 검정 수행
chi_square, p_value, df, expected = chi2_contingency(data)

#expected는 expected frequency

print( chi_square, p_value, df, expected)
```

#### 3.F검정

```python
from scipy import stats

# 두 표본 데이터
data1= [68, 56, 57, 54, 64, 48, 68, 56, 61, 58, 67, 49, 58, 58, 65, 54, 59, 55, 60, 62]
data2 =[65, 57, 57, 54, 64, 47, 67, 54, 60, 58, 65, 48, 57, 56, 64, 53, 57, 55, 61, 63]

# 두 표본 분산
variance1 = np.var(data1, ddof=1)
variance2 = np.var(data2, ddof=1)

# F-통계량 계산
f_value = variance1 / variance2

# 자유도
df1 = len(data1) - 1  # 분자의 자유도
df2 = len(data2) - 1  # 분모의 자유도

# p-값 계산
p_value = stats.f.sf(f_value, df1, df2)

print(p_value)
```
